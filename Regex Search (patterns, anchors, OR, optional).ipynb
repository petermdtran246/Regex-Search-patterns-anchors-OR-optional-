```python
# Regex Search (patterns, anchors, OR, optional)
```

```python
import re
```

```python
reviews = [
    "Anna helped me today!",
    "the service was awful",
    "amazing support from An",
    "He wanted a refund yesterday",
    "the item I needed arrived",
    "Absolutely loved the product",
    "I want to buy again"
]
```

```python
# 1️⃣ Find all reviews that start with the letter “A” or “a”.

a_reviews = []
patter_to_find = r"^(a|A)"
for i in reviews:
    if(re.search(patter_to_find, i)):
        a_reviews.append(i)
print(a_reviews)
```

```python
# Option for 1
a_reviews = []
patter_to_find = r"^[aA]"
for i in reviews:
    if(re.search(patter_to_find, i)):
        a_reviews.append(i)
print(a_reviews)
```

```python
# 2️⃣ Find all reviews that end with the letter “d”.

d_reviews = []
pattern_to_find = r"d$"
for i in reviews:
    if(re.search(pattern_to_find, i)):
        d_reviews.append(i)
print(d_reviews)
```

```python
# 3️⃣ Find all reviews that contain a form of “need” or “want”, including:
# need
# needed
# want
# wanted

all_reviews = []
pattern_to_find = r"(need|want)ed"
for i in reviews:
    if(re.search(pattern_to_find, i)):
        all_reviews.append(i)
print(all_reviews)
```

```python
# 4️⃣ Find all reviews that refer to Anna (both “Anna” and short form “An”).
    
Aa_reviews = []
pattern_to_find = r"(Anna|An)" 
for i in reviews:
    if(re.search(pattern_to_find, i, )):
        Aa_reviews.append(i)
print(Aa_reviews)
```

```python
# Optional for 4

Aa_reviews = []
pattern_to_find = r"(Ann?a?)"
for i in reviews:
    if(re.search(pattern_to_find, i, )):
        Aa_reviews.append(i)
print(Aa_reviews)
```
